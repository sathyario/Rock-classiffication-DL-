# Layers of Cnn

```python
classifier = tf.keras.models.Sequential(
    [
        tf.keras.layers.Conv2D(16,3,activation='relu',input_shape=(256,256,3)), 
        tf.keras.layers.MaxPool2D((2,2)),
        tf.keras.layers.Conv2D(8,2,activation='relu'),
        tf.keras.layers.MaxPool2D((2,2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128,activation = 'relu'),
        tf.keras.layers.Dense(7,activation = 'softmax')
    ]
)
```

# compile and summary

```python
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
classifier.summary()

### categorical_crossentropy is a loss function that is commonly used in multi-class classification problems.
```
